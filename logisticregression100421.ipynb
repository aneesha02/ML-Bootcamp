```python
#program to illustrate multiclass logistic regression
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
class LogReg:
    def __init__(self,alph,maxitr):
        self.alph=alph
        self.maxitr=maxitr
        
    #conversion of label values for the objective of multiclass logistic regression
    def convert(self,y,rowcnt,nclass): 
        y1=np.zeros((rowcnt,nclass))
        for i in range(0,nclass):
            for j in range(0,rowcnt):
                if y[j]==np.unique(y)[i]:
                    y1[j,i]=1
                else:
                    y1[j,i]=0
        return y1       
        
    #to calculate the predicted values of labels based on weights and independent variables 
    def predict(self,x): 
        ypred=np.zeros((x.shape[0],self.w.shape[1]))
        for i in range(0,self.w.shape[1]) :
            weight=self.w[:,i]
            h=weight[0]
            for j in range(1,colcnt):
                h+=weight[j]*x[:,j-1] 
            ypred[:,i]=h   
        return 1/(1+np.exp(-ypred))
    
    
    
    #to train the model. Starting with initial weights for each class of multiclass logistic regression, it calculates 
    #the initial predicted values of labels and corresponding error for all the classes. 
    #Gradient descent is performed over 'maxitr' number of iterations to find the new weights for each class.
    #We fix the optimum value of 'maxitr' by running the model by various trials.
    #We also assume the learning rate 'alph' at optimum level
    def fitting(self,x,y1,colcnt,nclass):
        self.w=np.random.randn(colcnt,nclass)
        print("Initial Weights=",self.w)
        ypred=self.predict(x)
        print("Initial predicted values of y=")
        err=y1-ypred
        print(ypred)
        print(err)
        for i in range(0,maxitr):
            for j in range(0,nclass):
                self.w[0,j]+=(alph)*np.sum(err[:,j]*ypred[:,j]*(1-ypred[:,j]))
                for k in range(1,colcnt):
                    self.w[k,j]+=(alph)*np.sum(x[:,k-1]*err[:,j]*ypred[:,j]*(1-ypred[:,j]))
            print("Calculated weights=")
            print(self.w)
            ypred=self.predict(x)
            print("Predicted values of y=")
            print(ypred)
            err=y1-ypred  
            print("Error=")
            print(err)
        print("Accuracy of train=",self.accuracy(y1,ypred)*100)
    #to test the accuracy of the predicted result
    def accuracy(self,y,ypred): 
        match=0
        for i in range(0,y.shape[1]):
            for j in range(0,y.shape[0]):
                if(y[j,i]==1 and ypred[j,i]>=0.5):
                    match+=1
        match=match/y.shape[0]
        return match
    
     
df=pd.read_csv('pythonlab\mnist_train_small.csv')
alph=0.0000001
maxitr=1000
nitr=0
rowcnt=len(df.index) #calculating the number of rows in the dataframe
colcnt=len(df.columns)#calculating the number of columns in the dataframe
x=(1/255)*np.array(df.iloc[:,1:colcnt])#normalizing the X values as X=(X-min)/(max-min)
y=np.array(df.iloc[:,0])
print(y)
nclass=len(np.unique(y)) #calculating the number of different classes present in y
#print(nclass)
#print(np.unique(y))
p=LogReg(alph,maxitr)
y1=p.convert(y,rowcnt,nclass)
print(y1)     
print("Number of rows=",rowcnt," Number of columns=",colcnt)
p.fitting(x,y1,colcnt,nclass)  


testdf=pd.read_csv('pythonlab\mnist_test.csv')
rowcnttest=len(testdf.index)
x2=(1/255)*np.array(testdf.iloc[:,1:colcnt])#normalizing the X values as X=(X-min)/(max-min)
y2=np.array(testdf.iloc[:,0])
nclass2=len(np.unique(y2))
y3=p.convert(y2,rowcnttest,nclass2)
ypred2=p.predict(x2)
print("Predicted values of y in test data")
print(ypred2)
#print(y3)
acc=p.accuracy(y3,ypred2)
print("Accuracy= ",acc*100,"%")
```
