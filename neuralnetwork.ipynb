```python
#program to implement neural network algorithm
import numpy as np
import pandas as pd

class smallneur:
    
    #calculates the value of the sigmoid activation function
    def sigmoid(self,x):
        return 1/(1+np.exp(-x))
    
    #calculates the value of the derivative of the sigmoid function
    def sig_der(self,x):
        return self.sigmoid(x)*(1-self.sigmoid(x))
    
    #calculates the value of the softmax activation function
    def softmax(self,x):
        ex = np.exp(x)
        return ex / ex.sum(axis=1, keepdims=True)
    
    #this method uses the sigmoid function and softmax function to calculate the activation values at hidden layer and output 
    #layer respectively. We are using 1 hidden layer consisting of 10 nodes. The cross entropy cost function has been used.
    # In the backpropagation the derivatives are calculated in 2 phases i.e output layer followed by hidden layer.
    #Hence the improved weights are calculated and this has been repeated over maxitr number of times.
    def fitting(self,x,y):
        #forpropagation
        rowcnt=x.shape[0]
        colcnt=x.shape[1]
        hidnode=100
        out_labels=10
        self.Wh=np.random.rand(colcnt,hidnode)
        self.bh=np.random.randn(hidnode)
        self.Wo=np.random.rand(hidnode,out_labels)
        self.bo=np.random.randn(out_labels)
    
        y1 = np.zeros((rowcnt, out_labels))
        for i in range(rowcnt):
            y1[i, y[i]] = 1
        
        flag=False
        nitr=0
        for nitr in range(maxitr):
            nitr+=1
            print('')
            print("Running Iteration number=",nitr)
            
            #phase 1
            zh=np.dot(x,self.Wh)+self.bh
            ah=self.sigmoid(zh)
            #phase 2
            zo=np.dot(ah,self.Wo)+self.bo
            ao=self.softmax(zo)
            (zo)
            print("Predicted values of y:")
            print(ao)
            
            #backpropagation
            
            #Phase 1
            outerr = ao - y1
            weight_err= ah
            dcost_wo = np.dot(weight_err.T, outerr)
            dcost_bo = outerr
            
            #Phase 2
            dzo_dah = self.Wo
            dcost_dah = np.dot(outerr , dzo_dah.T)
            dah_dzh = self.sig_der(zh)
            dzh_dwh = x
            dcost_wh = np.dot(dzh_dwh.T, dah_dzh * dcost_dah)
            dcost_bh = dcost_dah * dah_dzh
            
            #Updation         
            self.Wh-=alph*dcost_wh
            self.bh-=alph*dcost_bh.sum(axis=0)
            self.Wo-=alph*dcost_wo
            self.bo-=alph*dcost_bo.sum(axis=0)
            
            C=-np.sum(y1*np.log(ao))
            print("Loss function=",C)
            
            
        print("label values=")
        print(y1)
        print("Final predicted values=")
        print(ao)
        print("Accuracy of train=",self.accuracy(y1,ao)*100,"%")    
        
    #to calculate the predicted values of labels using weights and independent variables        
    def predict(self,x):
        zh=np.dot(x,self.Wh)+self.bh
        ah=self.sigmoid(zh)
       
        zo=np.dot(ah,self.Wo)+self.bo
        ao=self.softmax(zo)
        return ao
    
    #to calculate the accuracy of the model    
    def accuracy(self,y,ypred): 
        match=0
        for i in range(0,y.shape[1]):
            for j in range(0,y.shape[0]):
                
                if(y[j,i]==1 and ypred[j,i]>=0.5):
                    match+=1
        match=match/y.shape[0]
        return match
#training the model        
df=pd.read_csv('pythonlab\mnist_train_small.csv')
alph=0.0001
maxitr=100
thres=0.001
#nitr=0
rowcnt=len(df.index)
colcnt=len(df.columns)
x=(1/255)*np.array(df.iloc[:,1:colcnt])#normalizing the X values as X=(X-min)/(max-min)
#x=np.transpose(x)
y=np.array(df.iloc[:,0])
print("Number of rows=",rowcnt," Number of columns=",colcnt)
p=smallneur()
p.fitting(x,y)

#testing 
testdf=pd.read_csv('pythonlab\mnist_test.csv')
x2=(1/255)*np.array(testdf.iloc[:,1:colcnt])
y2=np.array(testdf.iloc[:,0])
y3= np.zeros((x2.shape[0], 10))
for i in range(x2.shape[0]):
        y3[i, y2[i]] = 1       
ypred2=p.predict(x2) 
print("Actual values of y=")
print(y3)
print("Predicted values of y in test:")
print(ypred2)
print("Accuracy of test=",p.accuracy(y3,ypred2)*100,"%")


```
